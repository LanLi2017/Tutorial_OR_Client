```python
# get column name
# parameters: project id
def GetColumnName(projectID):
    response=OR.get_models(projectID)
    column_model = response['columnModel']
    column_name = [column['name'] for column in column_model['columns']]
    return column_name

# project id: '2130630377229'
print GetColumnName('2130630377229')
```

# operations

Cell

```python
# text transform
# parameters: project id, column name, expression, on_error='set-to-blank', repeat= False, repeat_count=10
# returns:
'''
Lowercase column sponsor
'''
OR.text_transform('2130630377229','sponsor','value.toLowercase()')
```

```python
# cluster and edit
# parameters: project id, column name, clusterer type, function, params
# returns: clusters
clusters=OR.compute_clusters('2130630377229', 'event', 'binning','fingerprint')
clusters
```

```python
from_edit=OR.getFromValue(clusters)
to_edit=OR.getToValue(clusters)
default_edits=OrderedDict()
default_edits['fromBlank']='false'
default_edits['fromError']='false'
for f1,t1 in zip(from_edit, to_edit):
    default_edits['from']=f1
    default_edits['to']=t1
edits=[default_edits]
# mass edit
OR.mass_edit('2130630377229','event',edits)
```

```python
# single edit
# parameters: project id, row index, column index, input type, change value
# input type choices: number, boolean, date, default=String
# returns
OR.single_edit('2130630377229', 3, 3, 'string','tutorial_menu')
```
